```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_csv = "data/mouse_drug_data.csv"
clinical_csv = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_csv)
mouse_df.head()
clinical_df = pd.read_csv(clinical_csv)
clinical_df.head()

# Combine the data into a single dataset
drug_csv = pd.merge(clinical_df, mouse_df, on=['Mouse ID', 'Mouse ID'])

# Display the data table for preview
drug_csv.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_data = drug_csv.groupby(["Drug", "Timepoint"])
mean_tumor_data = grouped_data["Tumor Volume (mm3)"].mean()
mean_tumor_data

# Convert to DataFrame
tumor_df = pd.DataFrame(mean_tumor_data).reset_index()

# Preview DataFrame
tumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_tumor_error = grouped_data["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_error = pd.DataFrame(mean_tumor_error).reset_index()

# Preview DataFrame
tumor_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

munged_data = tumor_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
munged_data
```

```python
# Generate the Plot (with Error Bars)
Capomulin_error = tumor_error.loc[tumor_error["Drug"] == "Capomulin", "Tumor Volume (mm3)"]
Infubinol_error = tumor_error.loc[tumor_error["Drug"] == "Infubinol", "Tumor Volume (mm3)"]
Ketapril_error = tumor_error.loc[tumor_error["Drug"] == "Ketapril", "Tumor Volume (mm3)"]
Placebo_error = tumor_error.loc[tumor_error["Drug"] == "Placebo", "Tumor Volume (mm3)"]
Time = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
plt.errorbar(Time, munged_data["Capomulin"] , yerr= Capomulin_error, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(Time, munged_data["Infubinol"] , yerr= Infubinol_error, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(Time, munged_data["Ketapril"] , yerr= Ketapril_error, label= "Ketapril", marker= "D", color="black", linestyle='--')
plt.errorbar(Time, munged_data["Placebo"] , yerr= Placebo_error , label= "Placebo", marker= "s", color="green", linestyle='--')
plt.legend()
plt.title("Tumor Response to Treatment ")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()

# Save the Figure
plt.savefig("../Tumor_Response.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_ms_data = grouped_data["Metastatic Sites"].mean()
mean_ms_data

# Convert to DataFrame
ms_df = pd.DataFrame(mean_ms_data).reset_index()

# Preview DataFrame
ms_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_ms_error = grouped_data["Metastatic Sites"].sem()

# Convert to DataFrame
ms_error = pd.DataFrame(mean_ms_error).reset_index()

# Preview DataFrame
ms_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
munged_msdata = ms_df.pivot(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
munged_msdata
```

```python
# Generate the Plot (with Error Bars)
Capomulin_mserror = ms_error.loc[ms_error["Drug"] == "Capomulin", "Metastatic Sites"]
Infubinol_mserror = ms_error.loc[ms_error["Drug"] == "Infubinol", "Metastatic Sites"]
Ketapril_mserror = ms_error.loc[ms_error["Drug"] == "Ketapril", "Metastatic Sites"]
Placebo_mserror = ms_error.loc[ms_error["Drug"] == "Placebo", "Metastatic Sites"]
Time = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
plt.errorbar(Time, munged_msdata["Capomulin"] , yerr= Capomulin_mserror, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(Time, munged_msdata["Infubinol"] , yerr= Infubinol_mserror, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(Time, munged_msdata["Ketapril"] , yerr= Ketapril_mserror, label= "Ketapril", marker= "D", color="black", linestyle='--')
plt.errorbar(Time, munged_msdata["Placebo"] , yerr= Placebo_mserror , label= "Placebo", marker= "s", color="green", linestyle='--')
plt.legend()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.grid()

# Save the Figure
plt.savefig("../Metastatic_Spread.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice = grouped_data["Mouse ID"].count()
count_mice

# Convert to DataFrame
mice_df = pd.DataFrame(count_mice).reset_index()

# Preview DataFrame
mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
munged_mice = mice_df.pivot(index = "Timepoint", columns = "Drug", values = "Mouse ID")

# Preview the Data Frame
munged_mice
```

```python
# Generate the Plot (Accounting for percentages)
Time = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
plt.plot(Time, (munged_mice["Capomulin"]/25)*100, label = "Capomulin", marker= "o", color="red", linestyle="--")
plt.plot(Time, (munged_mice["Infubinol"]/25)*100 , label= "Infubinol", marker= "^", color="blue", linestyle="--")
plt.plot(Time, (munged_mice["Ketapril"]/25)*100 , label= "Ketapril", marker= "D", color="black", linestyle="--")
plt.plot(Time, (munged_mice["Placebo"]/25)*100 , label= "Placebo", marker= "s", color="green", linestyle="--")
plt.legend()
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%) ")
plt.grid()

# Save the Figure
plt.savefig("../Survival_Rate.png")

# Show the Figure
plt.show()
```
